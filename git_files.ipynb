```python
from bs4 import BeautifulSoup
```

```python
f = open('git.html', 'r', encoding='utf-8')
html_text = f.read()
f.close()
```

```python
soup = BeautifulSoup(html_text, 'html.parser')
```

```python
a_tags = soup.find_all("a", {'class':"html-attribute-value html-external-link"})
list_links = []
for each in a_tags:
    if 'blob' in each['href']:
        list_links.append(each['href'])
list_links
```

```python
markdown = ""
for each in list_links:
    markdown += '![alt text]' + '(' + each +')' + '\n'
markdown
```

```python
f = open('README.md', 'w')
f.write(markdown)
f.close()
```
